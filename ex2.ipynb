```python
# ex1
num_date = 100
rice_list = [1]
total_rice_list = [1]

for i in range(num_date-1):
    today_rice = rice_list[i] * 2
    rice_list += [today_rice]
    total_rice_list += [rice_list[i] + today_rice]

print(rice_list)
print(total_rice_list)
print(len(rice_list))
print(len(total_rice_list))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Rice num")
plt.xlabel("date")
plt.ylabel("rice num")
plt.plot(rice_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("Total rice num")
plt.xlabel("date")
plt.ylabel("total rice num")
plt.plot(total_rice_list)  # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
